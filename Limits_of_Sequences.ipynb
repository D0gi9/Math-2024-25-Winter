# Convergent vs. Divergent Sequences:

A sequence is convergent if it has a finite limit.
A sequence is divergent if it does not approach a finite value (e.g., it grows without bound or oscillates indefinitely).

# Important Properties:

Limits are unique.
If a sequence is bounded and monotonic, it always converges.
Common operations (e.g., addition, multiplication) on convergent sequences preserve convergence.

# Special Cases:

$\lim_{n \to \infty} \frac{1}{n} = 0$

$\lim_{n \to \infty} r^n = 0 \quad \text{for } |r| < 1$

Sequences without a finite limit may tend to infinity or oscillate (e.g.,$\lim_{n \to \infty} n = \infty$  
$\lim_{n \to \infty} (-1)^n$ diverges).

Calculate:


$\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}$

Solution:

Simplify by canceling $n^2$

$\frac{n^2 (1 + \frac{3}{n})}{n^2 (2 - \frac{2}{n})}$ = $\frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}$


Evaluate the limit as nâ†’âˆž

$\frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}$ $\to \frac{1 + 0}{2 - 0} = \frac{1}{2}$

**Final Answer**

$\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n} = \frac{1}{2}$

Find the limit of the sequence:

$a_n = \left(1 + \frac{1}{n}\right)^n$


We will show that:  $\lim_{n \to \infty} a_n = e$

Recall the definition of $e$

$e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n$


Formal verification using logarithms:

$\ln(a_n) = \ln\left(\left(1 + \frac{1}{n}\right)^n\right) = n \ln\left(1 + \frac{1}{n}\right)$

Let:

$b_n = n \ln\left(1 + \frac{1}{n}\right)$



Expand $\ln\left(1 + \frac{1}{n}\right)$ using a series approximation



$\ln\left(1 + \frac{1}{n}\right) \approx \frac{1}{n} - \frac{1}{2n^2} + \frac{1}{3n^3} - \dots$




Multiply by $n$:

$b_n = n\left(\frac{1}{n} - \frac{1}{2n^2} + \frac{1}{3n^3} - \dots\right) = 1 - \frac{1}{2n} + \frac{1}{3n^2} - \dots$



As $n \to \infty$, all higher-order terms $\left(-\frac{1}{2n}, \frac{1}{3n^2}, \dots\right)$ vanish:

$\lim_{n \to \infty} b_n = 1$


Exponentiate to find the limit of $a_n$



$\ln(a_n) = b_n$  and   $\lim_{n \to \infty} b_n = 1$



We have: $\lim_{n \to \infty} a_n = e^1 = e$.




**Final Answer:**



 $\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e$.
