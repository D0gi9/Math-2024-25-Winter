Extremum points in mathematics refer to the maximum and minimum values of a function. These points occur where the function changes direction, which is often at its critical points. Here's a concise explanation:


**Critical Points:**

A function $f(x)$ has critical points where its derivative $f'(x)$ equals zero or is undefined.


**types of Extremum:**


Local Maximum: $f(x)$ is greater than nearby values.

Local Minimum: $f(x)$ is less than nearby values.


First Derivative Test:


If $f‚Ä≤(x)$ changes from positive to negative at $c$, $f(c)$ is a local maximum.


If $f‚Ä≤(x)$ changes from negative to positive at $c$, $f(c)$ is a local minimum.



**Second Derivative Test:**


If $f‚Ä≤‚Ä≤(c)>0$, $f(c)$ is a local minimum.


If $f‚Ä≤‚Ä≤(c)<0$, $f(c)$ is a local maximum.


**Global Extremum:**


Compare all local extrema and boundary values to find the highest (maximum) or lowest (minimum) value on the entire domain.
