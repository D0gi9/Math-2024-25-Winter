# **Functions**

Functions in mathematics are relations that map each input to exactly one output. They are often written as f(x), where "f" is the name of the function and "x" is the input variable. For each value of "x," the function produces a unique output, often called "f(x)" or "y." Functions can be represented in various forms, such as equations, graphs, or tables.

Functions are widely used in algebra, calculus, and many other areas of mathematics. They can be classified into different types, including linear functions, quadratic functions, and exponential functions, based on the nature of their relationships. Functions are essential in describing how one quantity depends on another, making them fundamental in both mathematics and real-world applications.

In calculus, a function plays a central role in understanding how quantities change. Calculus primarily focuses on two key concepts related to functions: derivatives and integrals.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def h(x):
  return 1/x

# Create x values (avoiding x=0 where the function is undefined)
x = np.linspace(-10, 10, 400)
x = x[x != 0]  # Remove x=0 to avoid division by zero

# Calculate corresponding y values
y = h(x)

# Plot the function
plt.plot(x, y)
plt.title("Graph of h(x) = 1/x")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
  return np.exp(x)

def g(x):
  return np.log(x)  # np.log is the natural logarithm (ln)

# Create x values for f(g(x)) (x > 0 for ln(x) to be defined)
x_fg = np.linspace(0.1, 5, 400)

# Calculate f(g(x))
y_fg = f(g(x_fg))

# Create x values for g(f(x)) (all real numbers)
x_gf = np.linspace(-2, 2, 400)

# Calculate g(f(x))
y_gf = g(f(x_gf))

# Plot f(g(x))
plt.plot(x_fg, y_fg, label="f(g(x))")

# Plot g(f(x))
plt.plot(x_gf, y_gf, label="g(f(x))")

# Plot the line y = x for comparison (identity function)
plt.plot(x_fg, x_fg, linestyle="--", color="gray", label="y = x")

plt.title("Graphs of f(g(x)) and g(f(x))")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
  return x - 1

def f_inverse(x):
  return x + 1

# Create x values
x = np.linspace(-5, 5, 400)

# Calculate y values for f(x) and f‚Åª¬π(x)
y_f = f(x)
y_finv = f_inverse(x)

# Plot f(x)
plt.plot(x, y_f, label="f(x) = x - 1")

# Plot f‚Åª¬π(x)
plt.plot(x, y_finv, label="f‚Åª¬π(x) = x + 1")

# Plot the line y = x for comparison (reflection line)
plt.plot(x, x, linestyle="--", color="gray", label="y = x")

plt.title("Graph of f(x) and its Inverse")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
