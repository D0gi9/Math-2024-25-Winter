# LINEAR EQUATIONS


**Linear Equations: A Brief Overview**


A linear equation is a mathematical statement that represents a straight line when graphed. It is typically written in the form:

$y=mx+c$

Where:


$y$ is the dependent variable,

$x$ is the independent variable,

$m$ is the slope of the line (rate of change),

$c$ is the y-intercept (where the line crosses the y-axis).


Key Characteristics:

Degree: The highest power of the variable is 1.


Graph: The graph of a linear equation is always a straight line.


Solutions: The solution is the point(s) where the line intersects other lines or axes.

Types:


One variable:

$ax+b=0$


Two variables:

$ax+by+c=0$

**Why can't the following system of equations be solved using Cramer's rule?**



$x_1+2x_2+3x_3=3$ , $4x_1+5x-2+6x_3=2$ , $7x_1+8x_2+9x_3=1$


Step 1: Represent the system in matrix form

The system of equations can be written as:


$\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}$


Here, the coefficient matrix is:

$
A = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}$


Step 2: Determine if Cramer's rule is applicable


Cramer's rule can only be used to solve a system of linear equations if the determinant of the coefficient matrix $A$ is non-zero. If the determinant is zero, the system either has no solution or infinitely many solutions, and Cramer's rule cannot be applied.

Let’s calculate the determinant of matrix $A$:



$\det(A) = \begin{vmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{vmatrix}$


Using cofactor expansion along the first row:


$
\det(A) = 1 \cdot \begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} - 2 \cdot \begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} + 3 \cdot \begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix}$


Now, compute the 2x2 determinants:


$\begin{vmatrix} 5 & 6 \\ 8 & 9 \end{vmatrix} = (5)(9) - (6)(8) = 45 - 48 = -3$


$\begin{vmatrix} 4 & 6 \\ 7 & 9 \end{vmatrix} = (4)(9) - (6)(7) = 36 - 42 = -6$


$\begin{vmatrix} 4 & 5 \\ 7 & 8 \end{vmatrix} = (4)(8) - (5)(7) = 32 - 35 = -3$


Now substitute these values back into the determinant expression:


$det(A)=1⋅(−3)−2⋅(−6)+3⋅(−3)$


$det(A)=−3+12−9=0$


Step 3: Conclusion

The determinant of the coefficient matrix

$A$ is zero. Since the determinant is zero, the system of equations does not have a unique solution (it could either have no solution or infinitely many solutions). Therefore, Cramer's rule cannot be applied to solve this system.
