Vectors and Systems of Linear Equations


n the context of solving a system of linear equations using matrices, the coefficient matrix A plays a crucial role. If A is invertible (i.e., its determinant
det $(ùê¥)‚â†0$, we can solve the system using the inverse matrix method:



$x = A^{-1}b$


Here, $x$ is the solution vector, and $b$ is the right-hand side vector. However, if
$det(A)=0$, the matrix $A$ is singular, meaning it is not invertible. In such cases, the system may:

1)Have infinitely many solutions if the equations are dependent.
2)Be inconsistent if the equations contradict each other.

#Equations of lines on a plane


**Slope-Intercept Form:**

$y=mx+c$

$m$ is the slope (gradient) of the line, representing the rate of change of $y$ with respect to $x$.


$c$ is the y-intercept, the point where the line crosses the y-axis.

**Point-Slope Form:**


$y‚àíy_1=m(x‚àíx_1)$


This form uses a known point $(x_1,y_1)$ on the line and the slope $m$.


**General Form:**


$Ax+By+C=0$

In this form, $A,B,$ and $C$ are constants, and the equation can represent any line in a plane.

$Two-Point Form$

$y‚àíy_1$=$\frac{y_2 - y_1}{x_2 - x_1}$ $(x‚àíx_1‚Äã)$ This is used when two points (x_1,y_1) and $(x_2,y_2)$ on the line are known. The slope is calculated using these points.


**Important Concepts:**


Slope of a Line: The slope $m$ is the ratio of the vertical change to the horizontal change between two points on the line,


$m = \frac{y_2 - y_1}{x_2 - x_1}$


Perpendicular Lines:Two lines are perpendicular if the product of their slopes is $‚àí1,i.e.,ùëö_1 √ó ùëö_2= ‚àí1$

Parallel Lines:Two lines are parallel if they have the same slope,m_1=m_2$
