#Functions


**Definition:**


$A$ function $f$ maps an input $x$ from its domain $D$ to an output $y$ in its range $R$:


$y=f(x)$


**Types of Functions:**


**Linear Function:** $f(x)=mx+b$


$m$: slope, $b$: y-intercept.


**Quadratic Function:**

$f(x)=ax^2+bx+c$


**Exponential Function:**

$f(x)=a‚ãÖb^x$


**Trigonometric Function:**

$f(x)=sin(x),cos(x),tan(x)$


Key Properties:


Domain: Set of all possible inputs.

Range: Set of all possible outputs.

Increasing/Decreasing: Depends on the derivative , $f‚Ä≤(x)>0 or ùëì‚Ä≤(ùë•)<0$

Even/Odd Functions:

$f(‚àíx)=f(x) (even), f(‚àíx)=‚àíf(x) (odd).$

Common Operations:

$Composition: (f‚àòg)(x)=f(g(x))$

Inverse: If $y=f(x)$, the inverse
$ùëì^‚àí1(ùë¶) $satisfies  $ùëì(ùëì^‚àí1(ùë¶)) = y$

1)$f(x)=x^2$   , geogebra link:=https://www.geogebra.org/m/avjvcbke







```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
  return x**2

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the function
y = f(x)

# Create the plot
plt.plot(x, y)
plt.title("Graph of f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def g(x):
  return np.sqrt(x)

# Generate x values (starting from 0 to avoid errors with negative square roots)
x = np.linspace(0, 10, 100)

# Calculate y values using the function
y = g(x)

# Create the plot
plt.plot(x, y)
plt.title("Graph of g(x) = ‚àöx")
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)

# Show the plot
plt.show()
```

3)$h(x) = \frac{1}{x}$ ,  Geogebra:https://www.geogebra.org/m/scevtwt7




```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def h(x):
  return 1/x

# Generate x values, avoiding x=0 where the function is undefined
x = np.linspace(-10, 10, 100)
x = x[x != 0]  # Remove x=0 to avoid division by zero

# Calculate y values using the function
y = h(x)

# Create the plot
plt.plot(x, y)
plt.title("Graph of h(x) = 1/x")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid(True)

# Show the plot
plt.show()
```

**Functions:**

Geogebra link: https://www.geogebra.org/calculator/e9g4psnx



```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
  return x - 1

def f_inverse(x):
  return x + 1

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y1 = f(x)
y2 = f_inverse(x)
y3 = x  # For the line y = x

# Create the plot
plt.figure(figsize=(8, 6))

plt.plot(x, y1, label="f(x) = x - 1")
plt.plot(x, y2, label="f‚Åª¬π(x) = x + 1")
plt.plot(x, y3, label="y = x", linestyle="--")  # Dashed line for y = x

# Add labels, title, and legend
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphs of f(x), f‚Åª¬π(x), and y = x")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
