**Differential Equations Overview**


A differential equation (DE) is an equation that involves an unknown function and its derivatives. It can be classified as:


Types of Differential Equations

Ordinary Differential Equations (ODEs): Involves derivatives with respect to one variable.


$F(x, y, y', y'', \dots, y^{(n)}) = 0$

Partial Differential Equations (PDEs): Involves partial derivatives with respect to multiple variables.


$F(x, y, u, u_x, u_y, u_{xx}, u_{xy}, \dots) = 0$


Order: Highest derivative in the equation.

Degree: Power of the highest derivative (if polynomial).


**Solutions & Methods**


Separable Equations:

$\frac{dx}{dy} = g(x)h(y)$ $‚áí$ $\int \frac{1}{h(y)} dy = \int g(x) dx$


Linear First-Order ODE:


$\frac{dy}{dx} + P(x)y = Q(x)$


Solution:

$ye^{\int P(x) dx} = \int Q(x)e^{\int P(x) dx} dx + C$


Homogeneous & Non-Homogeneous Equations


$a_n y^{(n)} + a_{n-1} y^{(n-1)} + \dots + a_1 y' + a_0 y = f(x)$


If $f(x)=0$, it's homogeneous, otherwise non-homogeneous.

First, we take the first derivative with respect to $t$:


$$\frac{\partial \psi}{\partial t} = -A\omega \sin(\omega t + kx)$$

$$\frac{\partial^2 \psi}{\partial t^2} = -A\omega^2 \cos(\omega t + kx)$$

Step 2: Compute $\frac{\partial^2 \psi}{\partial x^2}$

First, we take the first derivative with respect to $x$:

$$\frac{\partial \psi}{\partial x} = -Ak \sin(\omega t + kx)$$

Taking the second derivative:

$$\frac{\partial^2 \psi}{\partial x^2} = -Ak^2 \cos(\omega t + kx)$$

Step 3: Check the Wave Equation

Substituting these derivatives into the wave equation:

$$(-A\omega^2 \cos(\omega t + kx)) - v^2 (-Ak^2 \cos(\omega t + kx)) = 0$$


Simplifying:

$$-A\omega^2 \cos(\omega t + kx) + v^2 Ak^2 \cos(\omega t + kx) = 0$$

$$A\cos(\omega t + kx)(-\omega^2 + v^2 k^2) = 0$$

$$-\omega^2 + v^2 k^2 = 0$$

$$\omega^2 = v^2 k^2$$

$$\omega^2 = (\frac{\omega}{k})^2 k^2 = \omega^2$$
